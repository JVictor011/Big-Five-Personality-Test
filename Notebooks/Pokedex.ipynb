```python
dfc = df.copy()
```

**Criação do Data Frame "dfc_c" com todas as linhas de "dfc" em que a coluna nome contem 'Mega'**

```python
dfc_c = dfc.copy()
dfc_c = dfc[dfc['Name'].str.contains(" Mega ")]
```

**Formatação dos nomes no Data Frame "dfc" nas linhas em que a coluna nome contem "Mega" transformando de "< nome > Mega < nome >" para "Mega < nome >"**

```python
for index, row in dfc.iterrows():
  if index in dfc_c.index:
    name = dfc.loc[index, 'Name'].split()
    name_formated = ' '.join(name[1:])
    dfc.loc[index, 'Name'] = name_formated
dfc
```

Gráfico de dispersão (scatter plot) para visualização da diferença de Attack e Speed entre os tipos de Pokémon Grass Poison e Fire.

```python
vec_attack = dfc.loc[dfc['Types'].isin(['Grass Poison', 'Fire']), 'Attack'].tolist()
vec_speed = dfc.loc[dfc['Types'].isin(['Grass Poison', 'Fire']), 'Speed'].tolist()

vec_names = []
for index, row in dfc.iterrows():
    if row['Types'] == 'Grass Poison' or row['Types'] == 'Fire':
        vec_names.append(row['Name'])

color_map = {name: plt.colormaps.get_cmap('hsv')(i / len(vec_names)) for i, name in enumerate(vec_names)}

scatter = plt.scatter(vec_attack, vec_speed, c=[color_map[name] for name in vec_names])

legend_elements = [plt.Line2D([0], [0], marker='o', color='w', label=name, markerfacecolor=color_map[name])
                   for name in vec_names]
plt.legend(handles=legend_elements, loc='upper left', bbox_to_anchor=(1.02, 1))

plt.xlabel('Attack')
plt.ylabel('Speed')

plt.show()
```
